```python
import pandas as pd
import numpy as np
import calendar
```

```python
data = pd.read_csv("test.csv",header = 0)
df = pd.DataFrame(data)

outputlist = []

year = str(df["arrival_date_year"][0])
month = str(list(calendar.month_name).index(df["arrival_date_month"][0]))
if len(month) == 1:
    month = '0'+month
day = str(df["arrival_date_day_of_month"][0])
if len(day) == 1:
    day = '0'+day
            
date = year+'-'+month+'-'+day

hotel = {'Resort Hotel': 0, 'City Hotel':0}
oneday = []

for i in range(len(df)):
    if (1):
    #str(df['is_canceled'][i]) == '0':
    #    continue
    #else:
        year = str(df["arrival_date_year"][i])
        month = str(list(calendar.month_name).index(df["arrival_date_month"][i]))
        if len(month) == 1:
            month = '0'+month
        day = str(df["arrival_date_day_of_month"][i])
        if len(day) == 1:
            day = '0'+day
            
        newdate = year+'-'+month+'-'+day
        
        if newdate != date:
            oneday2 = [date,hotel['Resort Hotel'], hotel['City Hotel']]
            res = [sum(x) / len(x) for x in zip(*oneday)]
            for r in res:
                oneday2.append(r)
            outputlist.append(oneday2)
            print(oneday2)
            hotel = {'Resort Hotel': 0, 'City Hotel':0}
            oneday = []
            date = newdate

        hotel[str(df["hotel"][i])] += 1
            
        lead_time = float(df["lead_time"][i])
        arrival_date_week_number = float(df["arrival_date_week_number"][i])
        stays_nights = float(df["stays_in_weekend_nights"][i]) + float(df["stays_in_week_nights"][i])
        adults = float(df["adults"][i])
        children = float(df["children"][i])
        babies = float(df["babies"][i])
            
        oneday.append([lead_time , arrival_date_week_number, stays_nights, adults, children, babies] )
        
oneday2 = [date,hotel['Resort Hotel'], hotel['City Hotel']]
res = [sum(x) / len(x) for x in zip(*oneday)]
for r in res:
    oneday2.append(r)
outputlist.append(oneday2)
print(oneday2)
```

```python
import pandas as pd
df = pd.DataFrame(outputlist, columns =['arrival_date','hotel_Resort','hotel_City',"lead_time" , "arrival_date_week_number", "stays_nights","adults", "children", "babies"]  )
```
