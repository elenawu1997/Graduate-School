```python
import pandas as pd
import numpy as np
import calendar
```

```python
data = pd.read_csv("train.csv",header = 0)
df = pd.DataFrame(data)
data2 = pd.read_csv("train_label.csv",header = 0)
df2 = pd.DataFrame(data2)


outputlist = []

year = str(df["arrival_date_year"][0])
month = str(list(calendar.month_name).index(df["arrival_date_month"][0]))
if len(month) == 1:
    month = '0'+month
day = str(df["arrival_date_day_of_month"][0])
if len(day) == 1:
    day = '0'+day
            
date = year+'-'+month+'-'+day

daily_revenue = 0
j = 0
for i in range(len(df)):
    if str(df['is_canceled'][i]) == '0':
        year = str(df["arrival_date_year"][i])
        month = str(list(calendar.month_name).index(df["arrival_date_month"][i]))
        if len(month) == 1:
            month = '0'+month
        day = str(df["arrival_date_day_of_month"][i])
        if len(day) == 1:
            day = '0'+day
            
        newdate = year+'-'+month+'-'+day
        
        if newdate != date:
            outputlist.append([date,daily_revenue, str(df2["label"][j])])
            print([date,daily_revenue, str(df2["label"][j])])
            
            date = newdate
            j += 1
            daily_revenue = 0
        
        stays_nights = float(df["stays_in_weekend_nights"][i]) + float(df["stays_in_week_nights"][i])
        daily_revenue += stays_nights*float(df["adr"][i])
        
outputlist.append([date,daily_revenue, df2["label"][j]])
print([date,daily_revenue, str(df2["label"][j])])
```

```python
import pandas as pd
df = pd.DataFrame(outputlist, columns =['arrival_date','daily revenue','label']  )
```

```python
df.to_csv("daily revenue and y.csv",index = False, columns =['arrival_date','daily revenue','label'] )
```
