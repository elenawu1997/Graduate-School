```python
import numpy as np
```

```python
def hat(X):
    
    T=np.mat(X.T)
    H = X.dot(np.linalg.inv(T.dot(X))).dot(T)
    
    return H
```

```python
X = np.mat([
    [1,-1,-2,-3,1],
    [1,-3,-2,-1,1],
    [1,-1,-1,-4,1],
    [1,0,-5,-1,5],
    [1,-2,-2,-2,1],
])
print(X)
print(X.T.dot(X))
print(np.linalg.inv(X.T.dot(X)))
hat(X)
```

```python
X2 = X*2
hat(X2)
```

```python
X3 = np.mat([
    [1,-1*2,-2*3,-3*4,1*5],
    [1,-3*2,-2*3,-1*4,1*5],
    [1,-1*2,-1*3,-4*4,1*5],
    [1,0*2,-5*3,-1*4,5*5],
    [1,-2*2,-2*3,-2*4,1*5],
])
hat(X3)
```
