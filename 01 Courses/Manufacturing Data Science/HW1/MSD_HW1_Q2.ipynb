```python
import pandas as pd
import numpy as np
import random
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model  import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import confusion_matrix
from sklearn.impute import KNNImputer
import statsmodels.api as sm
```

```python
#所有Header
Header=["age", "bp", "sg", "al", "su", "rbc", "pc", "pcc", "ba", "bgr",
        "bu", "sc", "sod", "pot", "hemo", "pcv", "wc", "rc", "htn", "dm",
        "cad", "appet", "pe", "ane", "class"]

#資料屬於category data的 Header
Header_cat=["sg", "al", "su", "rbc", "pc", "pcc", "ba", "htn", "dm","cad", "appet", "pe", "ane", "class"] 

df = pd.read_excel ('MDS_Assignment1_kidney.xlsx',header=None,names=Header)
```

```python
df
```

```python
for i in range(0,len(df)):
    if df['class'][i]!= 'ckd' and df['class'][i]!= 'notckd':
        df.drop([i],inplace=True)
df = df.reset_index(drop=True)
```

```python
df
```

```python
df_des = df.replace({'?':None})
des=df_des.describe()
```

```python
des
```

```python
# # of missing value by counting null
count_missing = df_des.isnull().sum(axis=0)
des= des.append(pd.Series(count_missing, name='# of missing value'))

# # of outlier by value out of (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)
Q1 = df_des.quantile(0.25)
Q3 = df_des.quantile(0.75)
IQR = Q3 - Q1
count_outlier = ((df_des < (Q1 - 1.5 * IQR)) | (df_des > (Q3 + 1.5 * IQR))).sum()
des= des.append(pd.Series(count_outlier, name='# of outlier'))
```

```python
des
```

```python
df = df.replace({'ckd':1, 'notckd':0})

y = df[Header[-1]]
X = pd.DataFrame(df[Header[0:-1]], columns = df[Header[0:-1]].columns)
```

```python
imp = SimpleImputer(missing_values = '?' , strategy = 'most_frequent')
X2 = pd.DataFrame( imp.fit_transform(X[Header_cat[:-1]]), 
                  columns = X[Header_cat[:-1]].columns)

for i in range(len(Header_cat)-1):
    X[Header_cat[i]] = X2[Header_cat[i]]
```

```python
X = pd.get_dummies(X, columns = Header_cat[:-1])
```

```python
X = X.replace({'?': None})

imputer = KNNImputer(n_neighbors=10, weights="uniform")
X2 = imputer.fit_transform(X)
X = pd.DataFrame(X2, columns = X.columns)
```

```python
pd.DataFrame(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random.randint(0, 1000))
```

```python
logreg = LogisticRegression(max_iter=10000)
rfe = RFE(logreg, 15)
rfe = rfe.fit(X_train, y_train.values.ravel())

par=[]
for i in range(len(rfe.ranking_)):
    if rfe.ranking_[i]==1:
        par.append(X.columns[i])
```

```python
print(par)
```

```python
print(par)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train[par], y_train)
```

```python
y_pred = logreg.predict(X_test[par])
print('Accuracy rate: {:.2f}'.format(logreg.score(X_test[par], y_test)))
```

```python
con = confusion_matrix(y_test, y_pred)
print(con)
```
