```python
import pandas as pd
```

```python
data = pd.read_excel("人口資料合併(百分比).xlsx",header= 0)
shop = pd.read_csv("里300_after.csv",header = 0)
```

```python
li = []
for d in data["區域里"]:
    li.append(d)
```

```python
data.iloc[l]
```

```python
lidata = []
for s in shop["完整里"]:
    s = s.replace("台","臺")
    l = li.index(s)
    lidata.append(data.iloc[l])
    print(len(lidata))
```
