```python
import os
import pandas as pd
import numpy as np
import re
from datetime import datetime
from bs4 import BeautifulSoup

import nltk
from nltk.stem import PorterStemmer
import string
```

# QA to one file

```python
files = [f for f in os.listdir(os.getcwd()+'/Download Data') ]
files = [f for f in files if 'csv' in f]
files_Q = [f for f in files if 'Q20' in f]
files_A = [f for f in files if 'A20' in f]
```

```python
files_Q = sorted(files_Q)
files_A = sorted(files_A)
```

```python
for i, f in enumerate(files_Q):
    print(f)
    if i == 0:
        df_Q = pd.read_csv(f)
    else:
        df_temp = pd.read_csv(f)
        df_Q = pd.concat([df_Q,df_temp],axis=0, ignore_index=True)
df_Q
```

```python
for i, f in enumerate(files_A):
    if i == 0:
        df_A = pd.read_csv(f)
    else:
        df_temp = pd.read_csv(f)
        df_A = pd.concat([df_A,df_temp],axis=0, ignore_index=True)
df_A
```

```python
datelist = pd.date_range(start="2018-08-02",end="2021-06-30").tolist()
datelist = [d.strftime("%Y-%m-%d") for d in datelist]
datelist
```

```python
QD = df_Q["CreationDate"].tolist()
QD = [d[:10] for d in QD]
AD = df_A["CreationDate"].tolist()
AD = [d[:10] for d in AD]
```

```python
for d in datelist:
    if d not in QD or d not in AD:
        print(d)
```

```python
df_Q.to_csv('Q_raw.csv', index = False)
df_A.to_csv('A.csv', index = False)
#df = pd.read_csv('A.csv')
#df
```
