```python
import os
import pandas as pd
import numpy as np
import re
from datetime import datetime
from bs4 import BeautifulSoup

import nltk
from nltk.stem import PorterStemmer
import string
```

# Q add AA time

```python
df_Q = pd.read_csv('Q_raw.csv')
df_A = pd.read_csv('A.csv')
```

```python
df_Q["Time_AA"] = ""
```

```python
for i in range(len(df_Q)):
    if i%10000 == 9999:
        print(i)
        
    QId = df_Q["Id"][i]
    
    t = datetime.strptime(df_Q["CreationDate"][i], "%Y-%m-%d %H:%M:%S")
    
    #Title, Body, AcceptedAnswer
    AA = df_A[(df_A["Id"] == QId) & (df_A["AcceptedAnswerId"] == df_A["Id.1"])]["CreationDate"].tolist()
    if AA:
        t_AA = datetime.strptime(AA[0], "%Y-%m-%d %H:%M:%S")
        Time_AA = t_AA - t
        df_Q["Time_AA"][i] = Time_AA.total_seconds()
        
```

```python
df_Q.to_csv('Q_time.csv', index = False)
```
