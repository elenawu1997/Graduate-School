```python
import pandas as pd
import numpy as np
```

```python
df_pred = pd.read_csv("Q_pred50.csv")
df_B = pd.read_csv("B.csv")
```

```python
df_p = df_pred[(df_pred["CreationDate"]<'2021-01-01') & (df_pred["CreationDate"]>'2020-01-01')]
df_p.reset_index(drop = True, inplace = True)
df_p
```

```python
df_p ["Qrecord"] = ""
df_p ["Arecord"] = ""
df_p ["Urecord"] = ""
df_p ["Arate"] = ""
df_p ["Pyrecord"] = ""
df_p ["Pyrate"] = ""
```

```python
for i in range(len(df_p)):
    
    id = df_p["Id"][i]
    df_temp = df_B[df_B["Id"]==id]

    if str(df_p["AcceptedAnswerId"][i]) == 'nan':
        A, U = 0, 1
    else:
        A, U = 1, 0
        
    if not df_temp.empty:
        A += len(df_temp[df_temp["AcceptedAnswerId"].notnull()])
        U += len(df_temp[df_temp["AcceptedAnswerId"].isnull()])
    
    py = 1
    for j in df_temp.index:
        if 'python' in df_B["tags"][j]:
            py += 1
        
    df_p ["Qrecord"][i] = len(df_temp)+1
    df_p ["Arecord"][i] = A
    df_p ["Urecord"][i] = U
    df_p ["Arate"][i] = A/(A+U)
    df_p ["Pyrecord"][i] = py
    df_p ["Pyrate"][i] = py/(len(df_temp)+1)
```

```python
df_p
```

```python
df_p["Accepted"] = ""
for i in range(len(df_p)):
    if str(df_p["AcceptedAnswerId"][i]) == 'nan':
        df_p["Accepted"][i] =0
    else:
        df_p["Accepted"][i] =1
```

```python
datelist = pd.date_range(start="2020-01-01",end="2020-12-31").tolist()
datelist = [d.strftime("%Y-%m-%d") for d in datelist]
QD = df_B["CreationDate"].tolist()
QD = [d[:10] for d in QD]
for d in datelist:
    if d not in QD:
        print(d)
```

```python
df_p.to_csv('Q_50+B.csv', index = False)
```
