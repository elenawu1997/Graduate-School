```python
from scipy.stats import ttest_ind_from_stats

answered_n = len(df_median[df_median["Answered"]==1])
unanswered_n = len(df_median[df_median["Answered"]==0])

answered_mean = df_median[df_median["Answered"]==1].mean()
unanswered_mean = df_median[df_median["Answered"]==0].mean()
answered_std = df_median[df_median["Answered"]==1].std()
unanswered_std = df_median[df_median["Answered"]==0].std()
for i in range(len(df_median.columns)):
    print(df_median.columns[i])
    print(answered_mean[i], unanswered_mean[i])
    print(ttest_ind_from_stats (answered_mean[i], answered_std[i],answered_n , 
                                unanswered_mean[i], unanswered_std[i],unanswered_n ,\
                                equal_var=False))
    print('\n')
```
