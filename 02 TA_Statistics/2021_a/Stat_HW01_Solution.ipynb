### Exercise 2.15 (8%)

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
plt.rcParams["figure.dpi"] = 120

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 1),       # 1 points vertical offset
                    textcoords="offset points",
                    size=6,              
                    ha='center',va='bottom')
```

```python
#讀excel檔
df = pd.read_excel('Xr02-15.xlsx')

#看看資料
print(df)
```

```python
#將檔案內資料由大到小排序(可省略)
df = df.sort_values(by=['Consumption of oil (barrels per day)'],ascending=False)

#看看資料(已排序)
print(df)
```

```python
labels = df['Country']

#sizes數字過大不易在圖中表示，自行調整，記得注意單位
sizes = []
for num in df['Consumption of oil (barrels per day)']:
    sizes.append(int(num/10000))
```

```python
fig, ax = plt.subplots()
rects = ax.bar(labels, sizes, width=0.8, bottom=None, align='center')
plt.title('Average oil consumption per day of the top 20 oil-comsuming natrions') #記得改標題
plt.ylabel('Consumption of oil (10K Barrels/ day)') # Y軸 label 配合單位
plt.xlabel('Country')
plt.xticks(labels,rotation='vertical')   # X軸 laebl 顯示角度調整，避免字被擋到

autolabel(rects)
plt.show()
```

### Exercise 2.53 (10%) (圖8%，說明2%)

```python
df = pd.read_excel('Xr02-39.xlsx')

#看看資料
print(df)
```

```python
#crosstab:分群在前，選項在後
contb = pd.crosstab( df["Social views"],df["Party"])
print(contb)
```

```python
S_label = ("Liberal", "Moderate", "Conservative")
```

```python
x = np.arange(len(S_label))  # the label locations
width = 0.2                  # the width of the bars
fig, ax = plt.subplots()

#根據資料調整rects數，X軸上位置，以及label
rects1 = ax.bar(x - 1.2*width, contb[1], width, label = "Democrat")
rects2 = ax.bar(x, contb[2], width, label = "Independent")
rects3 = ax.bar(x + 1.2*width, contb[3], width, label = "Republican")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frequency')
ax.set_xlabel('Socail view')
ax.set_title('Frequency by Party and Social View')
ax.set_xticks(x)
ax.set_xticklabels(S_label)
ax.legend()

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
```
